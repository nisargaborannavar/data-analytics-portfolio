### DATA LOADING

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/Titanic-Dataset.csv', encoding='latin1')
df.head() 
```

### DATA CLEANING


```python
print(df.shape)
df.info()
```

### Observations:

- Dataset contains 891 rows and 12 columns
- Each row represents a passenger

- Age column has 177 missing values (moderate missing data)
- Cabin column has 687 missing values (very high missing data)
- Embarked column has 2 missing values (minimal missing data)

- Dataset contains both numerical and categorical data

- Cabin column may not be useful due to large number of missing values
- Age column will require handling of missing values

### Hanadling Null values 


```python

df['Age'] = df.groupby(['Pclass', 'Sex'])['Age'].transform(
    lambda x: x.fillna(x.median())
)
df.drop(columns=['Cabin'], inplace=True, errors='ignore')
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

print(df.shape)
df.info()
```
