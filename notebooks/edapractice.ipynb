# Week 2 Practice Supert Set Deep Dive


## Objective
Explore the Superstore dataset to understand sales, profit, and category trends. Clean nulls and fix data types before analysis.

## Data loading

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/Sample - Superstore.csv', encoding='latin1')
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df.head()
```

## Data Cleaning

```python
# Check shape and info
print(df.shape)
df.info()
```

### Observations
- 9994 rows, 21 columns
- No null values

## Analysis

### Profit by Sub-Category


```python
df.groupby('Sub-Category')['Profit'].sum().plot(kind='bar', color='yellow')
plt.title('Profit by Sub-Category')
plt.xlabel('Profit')
plt.ylabel('Sub-Category')
plt.tight_layout()
plt.show()
```

### Sales by Region


```python
df.groupby('Region')['Sales'].sum().plot(kind = 'bar' , color = 'green' )
plt.title('Sales per Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.tight_layout()
plt.show
```

### Relation between Discount and Profit

```python
plt.scatter(df['Discount'], df['Profit'] , alpha = 0.3 , color = 'red' )
plt.title('Relation Between Discount and Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

## Insights
##### - Copiers has the highest profit among the Sub - Catgeory
##### - West Region has the highest sales compared to other regions
##### - The more discount given, the more likely the order loses money. Orders with 0 discount are the most profitable. Discounts above 50% almost always result in a loss.
