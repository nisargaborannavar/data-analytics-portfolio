# Week 2 EDA — Superstore Sales Dataset

## Objective
Explore the Superstore dataset to understand sales, profit, and category trends. Clean nulls and fix data types before analysis.

## Data Loading

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/Sample - Superstore.csv', encoding='latin1')
df.head() 
```

## Data Cleaning

```python
# Check shape and info
print(df.shape)
df.info()
```

### Observations
- 9994 rows, 21 columns
- No null values
- Order Date and Ship Date are strings — need to convert to datetime

```python
# Fix date columns
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Confirm fix
print(df[['Order Date', 'Ship Date']].dtypes)
```

## Analysis

```python
# Chart 1 - Sales by Category
df.groupby('Category')['Sales'].sum().plot(kind='bar', color='steelblue')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

markdown### Chart 2 - Profit by Region

```python
# Chart 2 - Profit by Region
df.groupby('Region')['Profit'].sum().plot(kind='bar', color='coral')
plt.title('Total Profit by Region')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

### Chart 3 - Sales vs Profit Scatter

```python
# Chart 3 - Sales vs Profit
plt.scatter(df['Sales'], df['Profit'], alpha=0.3, color='purple')
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

## Insights
- Technology has the highest total sales among all categories
- The West region generates the most profit
- Sales and Profit have a weak positive relationship — high sales does not always mean high profit
